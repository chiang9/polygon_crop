# Crop Image Interactively
With **img_crop**, we are able to crop image **interactively**. 

**img_crop** simply takes the image numpy array.

```python
path = 'Data/rabbit.jpeg'
img = Image.open(path).convert("RGB")
img_array = numpy.asarray(img)

display(img)
```

```python
newIm = img_crop(img_array)
```

```python
img = Image.fromarray(newIm, 'RGB')
display(img)
```
